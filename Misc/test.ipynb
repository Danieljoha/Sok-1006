1. Finn likevektspris og -kvantum for alle verdier av $t$.        


```python

import sympy as sp
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Markdown


def demand(p):
    return (500-p)/3.2

def supply(p, t):
    return ((100+p)/4.3) + t

p, t = sp.symbols('p t', real = True, positive = True)

eq_1 = sp.Eq(demand(p), supply(p, t))

pris_eq = sp.solve(eq_1, p)[0]
kvantum_eq = demand(pris_eq)


p_eq_fn=sp.lambdify(
    (t,p), 
    pris_eq
)

#p_eq_fn(t,p)
p_eq_fn(0,0)
p_eq_fn_uten_avgift = p_eq_fn(0,0)
p_eq_fn_med_avgift = p_eq_fn(4,0)



x_eq_fn=sp.lambdify(
    (t,p), 
    kvantum_eq
)

avgiften = 4

x_eq_fn(t,p)
x_eq_fn_uten_avgift = x_eq_fn(0,0)
x_eq_fn_med_avgift = x_eq_fn(avgiften,0)

x_eq_fn_med_avgift

print(p_eq_fn_uten_avgift)
print(x_eq_fn_uten_avgift)
print(p_eq_fn_med_avgift)
print(x_eq_fn_med_avgift)


avgift = x_eq_fn_med_avgift-x_eq_fn_uten_avgift
print(avgift)
prod = avgiften-avgift
```

```python
import sympy as sp
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Markdown
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display

def demand(p):
    return (500-p)/3.2

def supply(p, t):
    return ((100+p)/4.3) + t

p, t = sp.symbols('p t', real=True, positive=True)

eq_1 = sp.Eq(demand(p), supply(p, t))

pris_eq = sp.solve(eq_1, p)[0]
kvantum_eq = demand(pris_eq)

p_eq_fn = sp.lambdify((t, p), pris_eq)
p_eq_fn_uten_avgift = p_eq_fn(0, 0)
p_eq_fn_med_avgift = p_eq_fn(4, 0)

x_eq_fn = sp.lambdify((t, p), kvantum_eq)

avgiften = 4
x_eq_fn(t, p)
x_eq_fn_uten_avgift = x_eq_fn(0, 0)
x_eq_fn_med_avgift = x_eq_fn(avgiften, 0)

avgift = x_eq_fn_med_avgift - x_eq_fn_uten_avgift
avgiften_range = np.arange(-10, 21, 1)

prod = avgiften - avgift

x_num = np.linspace(0.1, 500, 100)

fig = go.FigureWidget()

# create the initial plot
fig.add_trace(go.Scatter(x=x_num, y=demand(x_num), name='Etterspørsel'))
fig.add_trace(go.Scatter(x=x_num, y=supply(x_num, 4), name='Tilbud med avgift'))
fig.add_trace(go.Scatter(x=x_num, y=supply(x_num, 0), name='Tilbud'))

fig.update_layout(
    title='Markedslikevekt',
    xaxis_title='Kvantum (x)',
    yaxis_title='Pris (p)',
    xaxis=dict(range=[160, 260], autorange=False),
    yaxis=dict(range=[40, 110], autorange=False),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None,
                  {'frame': {'duration': 200, 'redraw': True},
                   'fromcurrent': True,
                   'transition': {'duration': 0}}
                  ]
        )]
    )]
)

# create the slider
slider = widgets.IntSlider(value=10,
                           min=0,
                           max=len(avgiften_range)-1,
                           step=1,
                           description='Avgift:',
                           readout_format='d')

# define the animation function
def animate(avgift_index):
    fig.data[1].y = supply(x_num, avgiften_range[avgift_index])
    fig.data[2].y = supply(x_num, 0)
    q2 = np.linspace(0, p_eq_fn_med_avgift, 100)
    fig.data[3].y = demand(q2)
    fig.data[4].y = np.full((100,), x_eq_fn_uten_avgift+avgift)
    fig.data[5].y = np.full((100,), x_eq_fn_med_avgift-avgiften)

# link the slider and the animation function
widgets.interact(animate, avgift_index=slider)

# display the slider and the plot
display(widgets.VBox([fig, slider]))
widgets.interact(animate, avgift_index=slider);
```
